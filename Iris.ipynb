# Encoding the Target column(string to int)

```python
import category_encoders as ce
```

```python
encoder = ce.OrdinalEncoder(cols=["variety"],
                           return_df=True,
                            mapping=[{'col':'variety',"mapping":{'Setosa':0,'Versicolor':1,
                                                                    'Virginica':2}}])
```

```python
df = encoder.fit_transform(df) 
df
```

# Making Feature and Target columns

```python
X = df.drop(columns='variety') #Features
```

```python
Y = df.variety #Target
```

```python
X
```

```python
Y
```

# Scaling the Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X_array = ss.fit_transform(X)
```

```python
X_array
```

# Test Train Split

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_array,Y,test_size=0.2,random_state=42)
```

```python
print("length of X : ",len(X))
print("length of X_train : ",len(X_train))
print("length of X_test : ",len(X_test))
```

# Applying the KNN

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train,Y_train)
```

```python
Y_pred = classifier.predict(X_test) #predicting
```

```python
print(Y_pred)
```

# Evaluating the model

```python
from sklearn.metrics import r2_score

accuracy = r2_score(Y_test,Y_pred)
accuracy
```

```python
classifier.score(X_test,Y_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Accuracy: {accuracy:.2f}")

classification_rep = classification_report(Y_test, Y_pred)
print("Classification Report:\n", classification_rep)
```
